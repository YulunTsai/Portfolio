```python
import pandas as pd
```

```python
df = pd.read_csv('pc_games_igdb.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['aggregated_rating'] = df['aggregated_rating'].fillna(df['aggregated_rating'].median())
df['aggregated_rating_count'] = df['aggregated_rating_count'].fillna(0)
df['genres'].apply(lambda x: x if isinstance(x, list) else [])
df['first_release_date'] = pd.to_datetime(df['first_release_date'], unit='s')
df = df.drop(['genres'], axis=1)
df['popularity_score'] = df['popularity_score'] * 1000000
df['id'] = df['id'].astype(str)
```

```python
df.info()
```

```python
game_type_dict = {
    0: "Main Game",
    1: "DLC / Add-on",
    2: "Expansion",
    3: "Bundle",
    4: "Standalone Expansion",
    5: "Mod",
    6: "Episode",
    7: "Season",
    8: "Remake",
    9: "Remaster",
    10: "Expanded Game",
    11: "Port",
    12: "Fork"
}
df['game_type'] = df['game_type'].map(game_type_dict)
```

```python
source_dict = {
    1: "Steam",
    121: "IGDB"
}

df['popularity_source_id'] = df['popularity_source_id'].map(source_dict)
```

```python
type_dict = {
    1: "Visits",
    2: "Want to Play",
    3: "Playing",
    4: "Played",
    5: "24hr Peak Players",
    6: "Postitive Reviews",
    7: "Negative Reviews",
    8: "Total Reviews",
    9: "Global Top Sellers",
    10: "Most Wishlisted Upcoming"
}

df['popularity_type_id'] = df['popularity_type_id'].map(type_dict)
```
